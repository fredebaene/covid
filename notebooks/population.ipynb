###### DATA ANALYSIS

```python
# total population
```

```python
data['MS_POPULATION'].sum()
```

```python
# total population by province
```

```python
data_by_province = data.groupby(by='TX_PROV_DESCR_NL', as_index=False)['MS_POPULATION'].sum()
data_by_province.rename(
    columns={'TX_PROV_DESCR_NL' : 'PROVINCE', 'MS_POPULATION' : 'POPULATION'},
    inplace=True)
data_by_province['PROVINCE'] = data_by_province['PROVINCE'].str.replace('Provincie ', '')
data_by_province['PROVINCE'] = data_by_province['PROVINCE'].str.replace('-', '')
data_by_province['PROVINCE'] = data_by_province['PROVINCE'].str.strip()
```

```python
data_by_province
```

```python
mapping_provinces = {
    'Luik' : 'Liège',
    'Namen' : 'Namur',
    'Henegouwen' : 'Hainaut',
    'WaalsBrabant' : 'BrabantWallon',
    'Luxemburg' : 'Luxembourg'
}
```

```python
for k, v in mapping_provinces.items():
    data_by_province.loc[data_by_province['PROVINCE'] == k, 'PROVINCE'] = v
```
