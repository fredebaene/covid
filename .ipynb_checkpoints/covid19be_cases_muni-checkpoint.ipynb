### COVID-19 - CONFIRMED CASES BY DATE AND MUNICIPALITY

###### IMPORTING LIBRARIES

```python
import pandas as pd
```

###### READING DATA

```python
data = pd.read_csv('./data/covid19be_cases_muni.csv')
```

###### EXPLORING DATA

```python
data.info()
```

```python
nan_count = [(i, data[i].isna().sum()) for i in data.columns]
nan_count
```

###### CLEANING DATA

```python
data.dropna(inplace=True)
```

```python
nan_count = [(i, data[i].isna().sum()) for i in data.columns]
nan_count
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data['DATE'] = pd.to_datetime(data['DATE'])
```

```python
data.info()
```

```python
data['CASES'].replace(to_replace='<5', value='4', inplace=True)
```

```python
data['CASES'] = data['CASES'].astype(int)
```

```python
data.info()
```
