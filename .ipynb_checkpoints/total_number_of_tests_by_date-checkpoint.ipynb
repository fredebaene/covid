##### pandas.Series.value_counts(normalize=False, sort=True, ascending=False, bins=None, dropna=True)

parameters
- normalize : False : return the absolute frequencies; True : return the relative frequencies
- sort : True : sort the output based on the frequencies; False : do not sort the output based on the frequencies
- ascending : False : sort in descending order; True : sort in ascending order
- bins : use this parameter to calculate the frequencies by buckets
- dropna : True : do not include observations for which the value is missing; False : include observations for which the observation is missing

return
- the (relative) frequencies by unique value

```python
data['PROVINCE'].value_counts(dropna=False)
```

```python
data['REGION'].value_counts(dropna=False)
```

##### pandas.DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

parameters
- axis : 0 : drop rows with missing values; 1 : drop columns with missing values
- how : any : drop axis when at least 1 np.nan is there; all : drop axis when all values are np.nan
- thresh : n : drop axis when there are at least n values are np.nan
- subset
- inplace : True : modify the given DataFrame object; False : return a new DataFrame object

return
- None : the existing DataFra

```python
data.dropna(axis=0, how='any', inplace=True) # python script
```

```python
data['REGION'].value_counts(dropna=False)
```

```python
data['PROVINCE'].value_counts(dropna=False)
```

pandas.DataFrame.groupby(by=None, axis=0, level=None, as_index=True, sort=True, group_keys=True, squeeze=<object object>, observed=False, dropna=True)

```python
data_by_region = data.groupby(by='REGION') # python script
```

```python
data_by_region_province = data.groupby(by=['REGION', 'PROVINCE']) # python script
```

```python
total_tests_by_region = data_by_region[['TESTS_ALL', 'TESTS_ALL_POS']].sum() # python script
```

```python
total_tests_by_region_province = data_by_region_province[['TESTS_ALL', 'TESTS_ALL_POS']].sum() # python script
```

```python
total_tests_by_region['PCT_TESTS_POS'] = (total_tests_by_region['TESTS_ALL_POS'] / total_tests_by_region['TESTS_ALL']) * 100 # python script
```

```python
total_tests_by_region_province['PCT_TESTS_POS'] = (total_tests_by_region_province['TESTS_ALL_POS'] / total_tests_by_region_province['TESTS_ALL']) * 100 # python script
```

```python
total_tests_by_region.reset_index(inplace=True) # python script
```

```python
total_tests_by_region
```

```python
total_tests_by_region_province.reset_index(inplace=True) # python script
```
